```python
import pandas as pd
import numpy as np
```

```python
crime_anal=pd.read_csv('../data/crime_anal_in_Seoul.csv',encoding='euc-kr',index_col='구별')
crime_anal.head()
```

```python
from sklearn import preprocessing
```

```python
col=['강간','강도','살인','절도','폭력']

x=crime_anal[col].values
min_max_scaler=preprocessing.MinMaxScaler()

x_scaled=min_max_scaler.fit_transform(x.astype(float))
crime_anal_norm=pd.DataFrame(x_scaled, columns=col, index=crime_anal.index)

col2=['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm[col2]=crime_anal[col2]

crime_anal_norm.head()
```

```python
result_CCTV=pd.read_csv('../data/CCTV_result.csv',encoding='euc-kr',index_col='구별')
```

```python
result_CCTV.head()
```

```python
crime_anal_norm[['인구수','CCTV']]=result_CCTV[['인구수','CCTV수']]
crime_anal_norm.head()
```

```python
col=['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm['검거']=np.sum(crime_anal_norm[col],axis=1)
crime_anal_norm.head()
```

```python
col=['강간','강도','살인','절도','폭력']
crime_anal_norm['범죄']=np.sum(crime_anal_norm[col],axis=1)
crime_anal_norm.head()
```

```python
crime_anal_norm.to_csv('../data/crime_anal_norm_in_Seoul.csv',sep=',',encoding='euc-kr')
```
