```python
crime_police=pd.read_csv('../data/crime_in_Seoul_include_gu_name.csv',encoding='euc-kr')

del crime_police['Unnamed: 0']
```

```python
crime_police.head()
```

```python
station_name=[]

for name in crime_police['관서명']:
    station_name.append('서울'+str(name)+'경찰서')
    
station_name
```

```python
station_address=[]
station_lat=[]
station_lng=[]

for name in station_name:
    tmp=gmaps.geocode(name,language='ko')
    station_address.append(tmp[0].get('formatted_address'))
    
    tmp_loc=tmp[0].get('geometry')
    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])
    
    print(name+'-->'+tmp[0].get('formatted_address'))
```

```python
crime_police['lat']=station_lat
crime_police['lng']=station_lng
```

```python
map=folium.Map(location=[37.5502, 122.982], zoom_start=11)

for n in crime_police.index:
    folium.Marker([crime_police['lat'][n],
                  crime_police['lng'][n]]).add_to(map)
map
```
