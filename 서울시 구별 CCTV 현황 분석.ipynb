```python
import pandas as pd
import numpy as np
```

```python
CCTV_Seoul=pd.read_csv('../data/CCTV_in_Seoul.csv',encoding='euc-kr')
CCTV_Seoul.head()
```

```python
CCTV_Seoul.columns
```

```python
CCTV_Seoul.columns[0]
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0]:'구별'},inplace=True)
CCTV_Seoul.head()
```

```python
pop_Seoul=pd.read_excel('../data/population_in_Seoul.xlsx',encoding='euc-kr')
pop_Seoul.head()
```

```python
pop_Seoul=pd.read_excel('../data/population_in_Seoul.xlsx',
                       header=1,
                       usecols='B,D,E,F,H',
                       encoding='euc-kr')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0]:'구별',
                         pop_Seoul.columns[1]:'총계'}, inplace=True)
pop_Seoul.head()
```

```python
CCTV_Seoul.head()
```

```python
CCTV_Seoul.sort_values(by='총계',ascending=False).head(5)
```

```python
pop_Seoul.head()
```

```python
pop_Seoul.drop([0],inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique()
```

```python
pop_Seoul['외국인비율']=pop_Seoul['등록외국인'] / pop_Seoul['총계'] * 100
pop_Seoul.head()
```

```python
data_result=pd.merge(CCTV_Seoul, pop_Seoul, on='구별')
data_result
```

```python
data_result.rename(columns={'총계_y':'인구수'}, inplace=True)
```

```python
data_result
```

```python
data_result.head(3)
```

```python
del data_result['2012년 이전']
```

```python
del data_result['2012년']
del data_result['2013년']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
del data_result['2019년']
del data_result['2020년']
del data_result['2021년']
```

```python
data_result.head(3)
```

```python
data_result.rename(columns={'총계_x':'CCTV수'},inplace=True)
```

```python
data_result.set_index('구별',inplace=True)
```

```python
data_result.head(3)
```

```python
data_result.info()
```

```python
data_result['CCTV수']=pd.to_numeric(data_result['CCTV수'])
```

```python
data_result['CCTV수']=data_result.CCTV수.str.replace(',','')
```

```python
data_result.info()
```

```python
np.corrcoef(data_result['CCTV수'],data_result['인구수'])
```

```python
np.corrcoef(data_result['외국인비율'],data_result['CCTV수'])
```
