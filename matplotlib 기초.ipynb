```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
t=np.arange(0,12,0.01)
y=np.sin(t)
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.grid()
plt.legend()
plt.show()
```
